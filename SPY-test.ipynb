```python
import csv
with open('spy.csv') as file:
    spy = csv.reader(file)
    next(spy)

    date = []
    price = []
    share =[]
    average = []
    invest = []
    percent = []
    
    for row in spy:
        date.append(row[0])
        current_price = float(row[1])
        price.append(current_price)
        share.append(0)
        average.append(0)
        invest.append(0)
        percent.append(0)
```

```python
share[0] = 1
average[0] = price[0]*share[0] 
invest[0] = price[0]*share[0] 
percent[0] = 1

import math
threshold = 0.01

for n in range(1, len(date)):
    if price[n] > average[n-1]*(1-threshold):
        share[n] = 1
    else:
        share[n] = 1 + math.ceil((average[n-1] - price[n]) / average[n-1] / threshold)        
#    share[n] = 1 + math.ceil(max((average[n-1] - price[n]), 0) / average[n-1] / threshold)    
    invest[n] = share[n]*price[n]
    average[n] = sum(invest) / sum(share)
    percent[n] = price[n] /  average[n]
        
print(share)
print(sum(share))
print(average)
print(percent)
print(max(percent))
print(min(percent))
print("total $ is: " + str(math.floor(sum(invest))))
```
